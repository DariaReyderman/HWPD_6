```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
df = pd.read_csv("sport_athletes.csv")
```

### Exercise 1

```python
sport_type = df.groupby('Sport Type').agg({
    'Weight(kg)' : 'min',
    'Height(cm)' : 'max',
    'Wins' : 'max'}).rename(columns={'Weight(kg)' : 'max_weight', 'Height(cm)' : 'max_height', 'Wins' : 'max_wins'})
display(sport_type)
```

### Exercise 2

```python
the_tallest = df.sort_values(['Sport Type', 'Height(cm)'], ascending=[True, False])\
                .groupby('Sport Type').head(2)
result = the_tallest[['Sport Type', 'Athlete Name', 'Height(cm)']]
display(result)
```

### Exercise 3

```python
df_a = df.groupby(['Country', 'Sport Type']).max()
display(df_a)
```

```python
df_b = df.groupby(['Country', 'Sport Type']).max().loc[('USA', 'Basketball')]
display(df_b)
```

```python
df_c = df.groupby(['Sport Type', 'Weight(kg)']).max().xs(level='Sport Type', key='Soccer')
display(df_c)
```

```python
df_d = df[df['Sport Type'].isin(['Tennis', 'Soccer'])].groupby(['Sport Type', 'Weight(kg)']).max()
display(df_d)
```

```python
df_e = df.groupby(['Country', 'Sport Type']).max().swaplevel().sort_index(level=['Sport Type', 'Country'])
display(df_e)
```

```python
df_f = df.groupby(['Wins', 'Height(cm)']).agg({'Wins' : ['max', 'min', 'mean', 'std'], 'Height(cm)' : ['max', 'min', 'mean', 'std']})
display(df_f)
```
